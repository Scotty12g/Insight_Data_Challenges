# Employee Retention Data Challenge

## Goal

Employee turn-over is a very costly problem for companies. The cost of replacing an employee can be large (>100K USD), taking into account both the sunk costs spent recruiting and signing the employee, the decrease in productivity while searchig for a replacement, as well as the costs of finding a new employee. Thus, improving employee retention is a good investment in any venture. In this data challenge, I was given a data set with info about employees focused on predicting when an employee might quit by understanding the main drivers of employee churn.

## Challenge Description

We were given data about all employees who joined multiple comnpnaies between 2011/01/24 to 2015/12/13. For each employee, we also know if they are still at the company as of 2015/12/13, or if they  quit. Beside that, we have general info about the employee, such as avg salary during thier tenure, department within the company, and years of experience prior to joining the company.

The individual goals of this challange are to:

* Estimate the number of employees at each company for each day, assuming that the headcount starts from zero on 2011/01/23.

* Determine the main factors driving employee churn, and how they make sense.

* Suggest additional data that could help explain employee churn

## Data

We have 1 table. The table is:

Columns:
* employee_id : id of the employee. Unique by employee per company company_id : company id.
* dept : employee dept
* seniority : number of yrs of work experience when hired
* salary: avg yearly salary of the employee during her tenure within the company join_date: when the employee joined the company, it can only be between 2011/01/24 and 2015/12/13
* quit_date: when the employee left her job (if she is still employed as of 2015/12/13, this field is NA)

Let's start by estiamting company size (# of employees) for any given dayt for any given company. We can do this itteratively by day, adding hires and subtracting quits.

```python
#I create a new dataframe with employee count by day
allDates = pd.date_range(start = datetime(2011,01,23), end = datetime(2015,12,13))
employeeCount_df = pd.DataFrame(list(product(allDates, employee_retention.company_id.unique())), columns=['Date', 'company_id'])

def getEmployeeCount(x):
    curretDate = x.Date
    curretCompany = x.company_id
    return employee_retention[(employee_retention['join_date'] <= curretDate) &
                           ((employee_retention['quit_date'] > curretDate) | 
                            pd.isnull(employee_retention['quit_date'])) &
                             (employee_retention['company_id'] == curretCompany)].unique_id.count()


EmployeeCount = employeeCount_df.apply(lambda x: getEmployeeCount(x), axis = 1)
employeeCount_df['EmployeeCount'] = EmployeeCount

```

Now lets start to understand what's driving churn in the system. Since we have multiple variables which are a mix of categorical and continuous, our observations of each employee span different periods of time, and becuase there are obviously differences among companies, we're going to eventually use a Cox Proportional Hazard model - a type of survival analysis eventually.

First, let's just look at the survivorship curve overall with a Kaplan Meier Fit. We also plot the median from the curve, shoing approximately 50% survivorship.

```python
employee_retention2=employee_retention

def fillTimeNA(x):
    if pd.isnull(x.quit_date):
        return (datetime(2015,12,13)  - x.join_date).days
    else:
        return (x.quit_date - x.join_date).days

churn_timeAll = employee_retention2.apply(lambda x: fillTimeNA(x), axis = 1)
employee_retention2['churn_timeAll'] = churn_timeAll


E = employee_retention2['churn_binary']
T = employee_retention2['churn_timeAll']
```

```python
plt.figure(figsize=(10, 10))
ax = plt.subplot(111)

companySize_cluster_0 = (employee_retention["companySize_cluster"] == 0)
kmf.fit(T[companySize_cluster_0], event_observed=E[companySize_cluster_0], label="companySize_cluster_0")
kmf.plot(ax=ax, ci_force_lines=True)
print "Median survival time of cluster_0:", kmf.median_

companySize_cluster_1 = (employee_retention["companySize_cluster"] == 1)
kmf.fit(T[companySize_cluster_1], event_observed=E[companySize_cluster_1], label="companySize_cluster_1")
kmf.plot(ax=ax, ci_force_lines=True)
print "Median survival time of cluster_1:", kmf.median_

companySize_cluster_2 = (employee_retention["companySize_cluster"] == 2)
kmf.fit(T[companySize_cluster_2], event_observed=E[companySize_cluster_2], label="companySize_cluster_2")
kmf.plot(ax=ax, ci_force_lines=True)
print "Median survival time of cluster_2:", kmf.median_

companySize_cluster_3 = (employee_retention["companySize_cluster"] == 3)
kmf.fit(T[companySize_cluster_3], event_observed=E[companySize_cluster_3], label="companySize_cluster_3")
kmf.plot(ax=ax, ci_force_lines=True)
print "Median survival time of cluster_3:", kmf.median_
print ""

results = logrank_test( T[companySize_cluster_0],T[companySize_cluster_1], E[companySize_cluster_0], E[companySize_cluster_1])
print "companySize_cluster_0 vs companySize_cluster_1 " + str(results.p_value*6<0.05)

results = logrank_test( T[companySize_cluster_0],T[companySize_cluster_2], E[companySize_cluster_0], E[companySize_cluster_2])
print "companySize_cluster_0 vs companySize_cluster_2 " + str(results.p_value*6<0.05)

results = logrank_test( T[companySize_cluster_0],T[companySize_cluster_3], E[companySize_cluster_0], E[companySize_cluster_3])
print "companySize_cluster_0 vs companySize_cluster_3 " + str(results.p_value*6<0.05)

results = logrank_test( T[companySize_cluster_1],T[companySize_cluster_2], E[companySize_cluster_1], E[companySize_cluster_2])
print "companySize_cluster_1 vs companySize_cluster_2 " + str(results.p_value*6<0.05)

results = logrank_test( T[companySize_cluster_1],T[companySize_cluster_3], E[companySize_cluster_1], E[companySize_cluster_3])
print "companySize_cluster_1 vs companySize_cluster_3 " + str(results.p_value*6<0.05)

results = logrank_test( T[companySize_cluster_2],T[companySize_cluster_3], E[companySize_cluster_2], E[companySize_cluster_3])
print "companySize_cluster_2 vs companySize_cluster_3 " + str(results.p_value*6<0.05)

plt.ylim(0, 1);
plt.title("Survival function of Employees by Company Size Cluster");
plt.show()
```

Let's dive more deeply into the survivorship curve and look at which factors most explain differences in churn with a Cox Proportional Hazard model as suggested before, as sort of an omnibus test.

We'll start by incliding the following variables to explain churn time:

* department
* seniority cluster (which incorporates seniority and salary at the moment)
* employee count (the number of emplopyees on staff when the person quit, which can help account for possible company saturation)
* quit date quarter

We'll need to nest the analysis by company to account for the interactive relationships between these variables and company. We'll also need to encode categorical variables into binary ones by column, and scale employee count and seniority cluster data. We'll treat seniority cluster as a continuous variable since increases and decreases in it relfect both increases and decreases in salary and seniority.

```python

numberAdded = employee_retention.groupby(['company_id','dept','join_date']).unique_id.count().groupby(level=[0]).cumsum()
numberSubtracted = employee_retention.groupby(['company_id','dept','quit_date']).unique_id.count().groupby(level=[0]).cumsum()

numberAdded = numberAdded.to_frame().reset_index().rename(index=str, columns={"unique_id": "numberAdded"})
numberSubtracted = numberSubtracted.to_frame().reset_index().rename(index=str, columns={"unique_id": "numberSubtracted"})

employeeCount_df = pd.merge(employeeCount_df, numberAdded, left_on = ['company_id','dept','Date'], right_on = ['company_id','dept','join_date'])
employeeCount_df = pd.merge(employeeCount_df, numberSubtracted, left_on = ['company_id','dept','Date'], right_on = ['company_id','dept','quit_date'])

employeeCount_df['EmployeeCount']=employeeCount_df['numberAdded']-employeeCount_df['numberSubtracted']
employeeCount_df = employeeCount_df[['company_id','dept','Date','EmployeeCount']]
```

```python
g = sns.FacetGrid(employeeCount_df, col="dept",hue="company_id",col_wrap=3,size=4)
g.map(plt.plot,"Date", "EmployeeCount" )
plt.show()
```
